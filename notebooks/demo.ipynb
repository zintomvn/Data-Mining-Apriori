## 2. Nạp Thư Viện Python (Pandas)
Khai báo đường dẫn trỏ tới các file mà thuật toán Apriori vừa xuất ra.

```python
import pandas as pd
import os

# Thiết lập thư mục và các đường dẫn tương ứng
csv_dir = "../docs/chapter_5"
fi_path = os.path.join(csv_dir, "frequent_itemsets.csv")
ar_path = os.path.join(csv_dir, "association_rules.csv")
stats_path = os.path.join(csv_dir, "statistics.csv")
```

## 3. Hiển thị Bảng Tập Phổ Biến (Frequent Itemsets)
Xem chi tiết top 15 danh sách các mặt hàng (hoặc rổ mặt hàng) xuất hiện thường xuyên nhất trong CSDL siêu thị.

```python
if os.path.exists(fi_path):
    df_fi = pd.read_csv(fi_path)
    # Sắp xếp tổng quan theo Support Count giảm dần để xem đồ nào mua nhiều nhất
    df_fi = df_fi.sort_values(by="support_count", ascending=False).reset_index(drop=True)
    display(df_fi.head(15))
else:
    print("⚠ Không tìm thấy file báo cáo. Hãy chạy Cell 1 trước bằng lệnh !julia.")
```

## 4. Hiển thị Bảng Luật Kết Hợp (Association Rules)
Hiển thị bảng **Top 15 luật được sắp xếp theo mức độ chặt chẽ (Lift)** để gợi ý chiến dịch Cross-Selling tự động.

```python
if os.path.exists(ar_path):
    df_ar = pd.read_csv(ar_path)
    
    # Điểm format cho bảng thêm đẹp trước khi display
    df_ar_styled = df_ar.head(15).style.background_gradient(cmap="Blues", subset=["lift"])
    display(df_ar_styled)
else:
    print("⚠ Không tìm thấy file báo cáo. Hãy chạy Cell 1 trước bằng lệnh !julia.")
```

## 5. Thống Kê Overview Dataset
Toàn cảnh về kích thước mẫu dữ liệu được đào xới.

```python
if os.path.exists(stats_path):
    df_stats = pd.read_csv(stats_path)
    display(df_stats.style.hide(axis="index"))
else:
    print("File statistics.csv không tồn tại.")
```
